## 1) Importing All Modules

```python
import cv2 as cv
import numpy as np
from PIL import Image
import imageio
from matplotlib import pyplot as plt, patches as patches


```

### Reading the Image Using the Pillow and then resizing the image

```python

Messi = Image.open('Messi.jpeg').convert('L')
Messi = Messi.resize((640, 640), Image.ANTIALIAS)

Messi=np.asarray(Messi)
print(Messi.shape)
plt.imshow(Messi,cmap = 'gray')
plt.title('Messi')
plt.show()


Ronaldo = Image.open('Ronaldo.jpeg').convert('L')
Ronaldo = Ronaldo.resize((640, 640), Image.ANTIALIAS)
Ronaldo=np.asarray(Ronaldo)
print(Messi.shape)

plt.imshow(Ronaldo,cmap = 'gray')
plt.title('Ronaldo')
plt.show()
```

### We First  compute the 2 - dimensional discrete Fourier Transform using the Fast Fourier Transform (FFT)
### fftshift then Shifts the zero-frequency component to the center of the spectrum. 
### To obtain the Phase Spectrum we use np.angle 
### to Obtrain the Magnitude spectrum we use the following formula 

```python
fourierMessi = np.fft.fft2(Messi)
fshift1 = np.fft.fftshift(fourierMessi)
phase_Messi = np.angle(fshift1)
magnitude_Messi = 20*np.log(np.abs(fshift1))
```

### Now we will Visualize the Phase Spectrum from first image

```python

plt.imshow(phase_Messi,cmap = 'gray')
plt.title("phase_Messi")
plt.show()
```

### Lets also Visualize the Magnitude Spectrum that we got from FFT


```python


plt.imshow(magnitude_Messi,cmap = 'gray')
plt.title("magnitude_Messi")
plt.show()
```

### Now we will also Apply the Above steps for our second pic , that is get the image's fourier transform, get the Phase and Magnitude spectrum 

```python
fourierRonaldo = np.fft.fft2(Ronaldo)
fshift2 = np.fft.fftshift(fourierRonaldo)
phase_Ronaldo = np.angle(fshift2)
magnitude__Ronaldo = 20*np.log(np.abs(fshift2))
```

### Now to Use The Magnitude of Image 1 and Phase Spectrum of 2nd Image , we use the Following Formula , to get the job done 

```python
Mag_Phase = np.multiply(magnitude_Messi, np.exp(1j*phase_Ronaldo))

#combined = np.multiply(np.abs(fourierMessi), np.exp(1j*np.angle(f2)))
Mag_Phase = np.multiply(np.abs(fourierMessi), np.exp(1j*np.angle(fourierRonaldo)))

ReverseImage = np.real(np.fft.ifft2(np.real(Mag_Phase)))
```

### Now we Visualize the Effect of mergin Magnitude and Spectrum of different Images 


```python
print(ReverseImage)

plt.imshow(ReverseImage,cmap = 'gray')
plt.title("Reconstructed image")
plt.show()
```
