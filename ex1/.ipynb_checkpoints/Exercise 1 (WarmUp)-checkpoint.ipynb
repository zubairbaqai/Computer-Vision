## 1) Importing All Modules

```python
from scipy import misc
import imageio
from PIL import Image
from matplotlib import pyplot as plt, patches as patches
from skimage.transform import rotate,rescale,resize
import numpy as np
from skimage import data
from skimage.util import img_as_ubyte
from skimage import io
from skimage.morphology import erosion, dilation, opening, closing , square
from skimage.morphology import disk
from skimage.filters import gaussian, threshold_otsu
from skimage import measure
from skimage.segmentation import clear_border
from skimage.color import label2rgb

```

### 2) Reading and writing Image using scipy

```python
#ReadMessi = misc.imread('Messi.jpeg') #### Deprecated
#misc.imsave('NewMessi.png', ReadMessi) ### Deprecated


ReadMessi = imageio.imread('Messi.jpeg')
imageio.imwrite('NewMessi.png', ReadMessi[:, :, 0]) ## Saving original image as GreyScale with taking only Red part of image 
```

### 3) Reading and writing image using Pillow

```python
image = Image.open("Messi.jpeg").convert('LA')
image.save("NewMessi.png") 
```

### 4)Displaying image using imshow() of matplotlib

```python
image = plt.imread('Messi.jpeg')
plt.imshow(image)
plt.title('Messi')
plt.show()
```

### 4) Cropping image using Slicing

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Uncroppedimage')
plt.subplot(1,2,2)
CroppedImage = image[10:100,30:120,:]
plt.imshow(CroppedImage)
plt.title('CroppedImage')
plt.show()
```

### 5) Geometric transformation (Applying Rescaling, Resizing and Rotation)

```python
Rescaling = rescale(image,2,anti_aliasing=False,multichannel=True)
Resizing = resize(image,(image.shape[0] // 4, image.shape[1] // 4),anti_aliasing=True)
Rotation = rotate(image, 90,resize=True)


plt.subplot(2,2,1),plt.imshow(image,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(Rescaling,cmap = 'gray')
plt.title('Rescaling'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(Resizing,cmap = 'gray')
plt.title('Resizing'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(np.real(Rotation),cmap = 'gray')
plt.title('rotate'), plt.xticks([]), plt.yticks([])

plt.show()
```

#### 6) Applying morphologies

```python
orig_phantom = img_as_ubyte(data.shepp_logan_phantom())
plt.imshow(orig_phantom)
plt.title('ImageOn Which we apply morphologies')
plt.show()

selem = disk(6) 


###Morphological erosion sets a pixel at (i, j) to the minimum over all pixels in the neighborhood centered at (i, j)
eroded = erosion(orig_phantom, selem)


##Morphological dilation sets a pixel at (i, j) to the maximum over all pixels in the neighborhood centered at (i, j). Dilation enlarges bright regions and shrinks dark regions.
dilated = dilation(orig_phantom, selem)


#Morphological opening on an image is defined as an erosion followed by a dilation. Opening can remove small bright spots (i.e. “salt”) and connect small dark cracks.
opened = opening(orig_phantom, selem)


#Morphological closing on an image is defined as a dilation followed by an erosion. Closing can remove small dark spots (i.e. “pepper”) and connect small bright cracks. To illustrate this more clearly, let’s add a small crack to the white border:
phantom = orig_phantom.copy()
phantom[10:30, 200:210] = 0

closed = closing(phantom, selem)
```

```python
plt.subplot(2,2,1),plt.imshow(closed,cmap = 'gray')
plt.title('closed'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(eroded,cmap = 'gray')
plt.title('eroded'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(dilated,cmap = 'gray')
plt.title('dilated'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(np.real(opened),cmap = 'gray')
plt.title('opened'), plt.xticks([]), plt.yticks([])

plt.show()
```

#### 7) Experimenting RegionProps

```python
coins_image = data.coins()


thres = threshold_otsu(coins_image)
closing = closing(coins_image > thres,square(5))
clearborder = clear_border(closing)
blurred = gaussian(clearborder, sigma=5)



binary = blurred > threshold_otsu(blurred)
labels = measure.label(binary)

plt.figure(figsize=(10,10))
plots = {'Original': coins_image, 'Blurred': blurred, 
         'Binary': binary, 'Labels': labels}
fig, ax = plt.subplots(1, len(plots))
for n, (title, img) in enumerate(plots.items()):
    cmap = plt.cm.gnuplot if n == len(plots) - 1 else plt.cm.gray
    ax[n].imshow(img, cmap=cmap)
    ax[n].axis('off')
    ax[n].set_title(title)
plt.show(fig)
```
